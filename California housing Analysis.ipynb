```python
#q10 Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.
```

```python
#  function to take each row as input for the value of coulmn total_bedrooms and apply method used to aply functio to each row of dataset
def get_total_bedroom_size(row):
    if row['total_bedrooms'] < 10:         
        return 'small'
    elif row['total_bedrooms'] <1000:
        return 'medium'
    else:
        return 'large'
df['total_bedrooms_size'] = df.apply(get_total_bedroom_size, axis=1)
print(df)
```
